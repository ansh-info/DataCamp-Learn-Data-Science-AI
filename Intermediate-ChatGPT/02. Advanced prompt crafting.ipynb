#### Negotiate a raise

Imagine you are preparing to ask for a raise at a workplace. You engage with ChatGPT in a simulated negotiation for a raise, utilizing techniques from Chris Voss's book "Never Split the Difference." This will help you practice and refine your negotiation skills using real-time feedback and techniques from an expert negotiator.

Here's a prompt to start the negotiation:

```
<context>
I am seeking a raise.
</context>
<task>
We will engage in a negotiation scenario. You will take two roles:
<role_manager>
As my manager, you will evaluate the request for a raise.
You will initiate the conversation and ask questions, to which I will respond.
</role_manager>
<role_ChrisVoss>
You will act as Chris Voss, the former FBI hostage negotiator.
After I have answered a question from my manager, provide a tip to help improve my response.
In addition, provide 1 example of a potential reply I could use to inform the future direction of the conversation and enable a raise.
</role_ChrisVoss>
As role_manager, you will continue asking questions after role_ChrisVoss has provided insight.
When I say "STOP," the negotiation will end.
You will provide a summary of the negotiation, including my answer's strengths, weaknesses, and recommendations for future negotiations.
</task>

Let's begin by asking me my first question, to which I will respond.

```

5\. 1. Persona
--------------

00:32 - 00:52

Adopting a persona ensures ChatGPT's output is specific, relevant, and useful. Instead of simply stating, "You are a writing expert," we add detail: it plays on human emotions, sparks intrigue, is counter-narrative, and makes people stop and think.

6\. 1. Persona
--------------

7\. 2. Framework
----------------

01:01 - 01:15

The framework determines the style of your response. It's like having bumpers up on your bowling lane so that the ball doesn't land in the gutter. A framework steers your prompt to get the output you deserve.

----------------

12\. 2. Framework - Positive future
-----------------------------------

01:41 - 01:49

Next, we paint a positive future. This is what a solution could look like and how it could make you feel.

13\. 2. Framework - Solution
----------------------------

01:49 - 02:01

The final step is to present the solution itself. We have now constructed our narrative arc and this completes our five-step PAIPS writing framework.

14\. 2. Framework

15\. 3. Specifications
----------------------

02:08 - 02:34

How does ChatGPT control the length and output format of our response? We direct this through specifications. We also want to infuse Hemingway's focus on action and dialogue to create clear, impactful, and succinct content that engages readers without unnecessary wording. You can adjust these specifications according to the context of your writing, whether it be an email, a social media post, or even a short story for a friend.

16\. 4. Example
---------------

02:34 - 02:59

The example we provide ChatGPT doesn't have to match the topic we want to write about, but it gives ChatGPT the framework to build its response from. This is called one-shot prompting. We're providing a model answer for ChatGPT to replicate. It's a bit like seeing the answers to a test and knowing how to meet the criteria.
